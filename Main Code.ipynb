### TRANSFORMING NUMERICAL FEATURES INTO BINS

```python
def get_bins(series):
    QUANTILES = [0,0.33,0.66,1]
    bins = []
    for quantile in QUANTILES:
        quantile = series.quantile(q=quantile)
        bins.append(quantile)
    return bins

def transform_numerical(df, column):
    to_bins = get_bins(df[column])
    for i in range(0,len(to_bins)-1):
        df.loc[(df[column] >= to_bins[i]) & (df[column] <= to_bins[i+1]), column] = i
    df[column] = df[column].astype("int")
numerical_columns = ["total_hours", "total_views", "total_followers", "record_viewers",
                "hours_streamed", "hours_watched", "average_viewers", "active_days", "average_hours",
                "active_days_per_week"]

for numerical in numerical_columns:
    transform_numerical(channels_df, str(numerical))
channels_df
```

```python
### CODE SNIPPET GET FROM THE AUTHOR OF THE PACKAGE, FROM STACKOVERL: https://stackoverflow.com/questions/49896816/how-do-i-optimize-the-hyperparameters-of-lightfm
### THIS WAS RUN IN ANOTHER COMPUTER WITH BETTER PERFORMANCE

import itertools

import numpy as np

def sample_hyperparameters():
    """
    Yield possible hyperparameter choices.
    """

    while True:
        yield {
            "no_components": np.random.randint(16, 64),
            "learning_rate": np.random.exponential(0.05),
            "num_epochs": np.random.randint(5, 50),
        }


def random_search(train, test, num_samples=15, num_threads=1):
    """
    Sample random hyperparameters, fit a LightFM model, and evaluate it
    on the test set.

    Parameters
    ----------

    train: np.float32 coo_matrix of shape [n_users, n_items]
        Training data.
    test: np.float32 coo_matrix of shape [n_users, n_items]
        Test data.
    num_samples: int, optional
        Number of hyperparameter choices to evaluate.


    Returns
    -------

    generator of (precision@6, hyperparameter dict, fitted model)

    """

    for hyperparams in itertools.islice(sample_hyperparameters(), num_samples):
        num_epochs = hyperparams.pop("num_epochs")

        model = LightFM(**hyperparams)
        model.fit(train, epochs=num_epochs, num_threads=num_threads)

        score = precision_at_k(model, test, train_interactions=train, k= 6, num_threads=num_threads).mean()

        hyperparams["num_epochs"] = num_epochs

        yield (score, hyperparams, model)


if __name__ == "__main__":

    (score, hyperparams, model) = max(random_search(train, test, num_threads=2), key=lambda x: x[0])

    print("Best score {} at {}".format(score, hyperparams))
```

```python
POPULAR_CHANNELS = list(df.following_to.value_counts().sort_values(ascending=False)[0:100].index)
```

```python
def precision_at_k_MPC_test(df, k):
    
    count = 0
    for user in df.username.unique():
        user_data = df[df.username==user]
        total_followed = len(user_data)
        
        hide_n = int(total_followed * 0.666)
        
        to_hide = list(user_data[:hide_n].following_to.values)
        to_test = list(user_data[hide_n:].following_to.values)
        
        clean = [channel for channel in POPULAR_CHANNELS if channel not in to_hide]
        
        intersection = [i for i in clean[0:k] if i in to_test]
        count += len(intersection)
    hit_result = count / k
    return hit_result/df.username.nunique()
```
